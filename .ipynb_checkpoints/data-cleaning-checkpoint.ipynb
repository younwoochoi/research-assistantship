```python
for i, row in video_df.iterrows():
    if pd.isnull(video_df.at[i, 'number_comments']):
        none = "N/A"
        video_df.at[i, 'number_comments'] = none
    elif type(video_df.at[i, 'number_comments']) is int:
        continue
    elif not(type(video_df.at[i, 'number_comments']) is int):
        new_input = float(video_df.at[i, 'number_comments'].strip("댓글개Comments").replace(",", ''))
        video_df.at[i, 'number_comments'] = new_input
```

```python
zero = 0
for i, row in video_df.iterrows():
    if pd.isnull(video_df.at[i, 'view_number']):
        none = 'N/A'
        video_df.at[i, 'view_number'] = none
    #elif type(video_df.at[i, 'view_number']) is int:
        #continue
    elif '없음' in video_df.at[i, 'view_number'] or "No" in video_df.at[i, 'view_number']:
        video_df.at[i, 'view_number'] = zero
    elif '현재' in video_df.at[i, 'view_number']:
        video_df.at[i, 'view_number'] = 'N/A'
    elif 'watching' in video_df.at[i, 'view_number']:
        video_df.at[i, 'view_number'] = 'N/A'
    elif 'waiting' in video_df.at[i, 'view_number']:
        video_df.at[i, 'view_number'] = 'N/A'
    elif '조회수' in video_df.at[i, 'view_number'] or 'view' in video_df.at[i, 'view_number']:
        new_input = float(video_df.at[i, 'view_number'].strip('조회수views').replace(',', ''))
        video_df.at[i, 'view_number'] = new_input
```

```python
for i, row in video_df.iterrows():
    if type(video_df.at[i, 'number_likes']) is int:
        continue    
    elif pd.isnull(video_df.at[i, 'number_likes']):
        none = 'N/A'
        video_df.at[i, 'number_likes'] = none
    elif type(video_df.at[i, 'number_likes']) is float:
        continue
    elif "천" in video_df.at[i, 'number_likes']:
        thousand = float(video_df.at[i, 'number_likes'].strip("천"))*1000
        video_df.at[i, 'number_likes'] = thousand
    elif "K" in video_df.at[i, 'number_likes']:
        thousand_eng = float(video_df.at[i, 'number_likes'].strip("K"))*1000
        video_df.at[i, 'number_likes'] = thousand_eng
    elif "만" in video_df.at[i, 'number_likes']:
        tenk_kor = float(video_df.at[i, 'number_likes'].strip("만"))*10000
        video_df.at[i, 'number_likes'] = tenk_kor
    elif "좋아요" in video_df.at[i, 'number_likes']:
        video_df.at[i, 'number_likes'] = 'N/A'
    elif "Like" in video_df.at[i, 'number_likes']:
        video_df.at[i, 'number_likes'] = 'N/A'
```
