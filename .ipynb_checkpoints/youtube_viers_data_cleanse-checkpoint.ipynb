```python
import pandas as pd
from pathlib import Path
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
path = Path('audience_overlap_complete_scrap_results_v1.xlsx')
```

```python
df = pd.read_excel(path)
```

```python
df
```

```python
id_lst = []
curr = ''
single_lst = []
channel_id = []
for i, row in df.iterrows():
    if curr != df.at[i, 'channel_name']:
        curr = df.at[i, 'channel_name']
        channel_id.append(curr)
        id_lst.append(single_lst)
        single_lst = []
    if not isinstance(df.at[i, 'commenters_id'], float):
        lst = df.at[i, 'commenters_id'].strip('[] ').replace("'", '').split(', ')
        single_lst.extend(lst)
    else:
        single_lst.extend([])
```

```python
commenter_set_lst = []

for channel in id_lst:
    commenter_set = set()
    for commenter in channel:
        commenter_set.add(commenter)
    commenter_set_lst.append(commenter_set)

len(commenter_set_lst)
```

```python
channel1 = []
channel2 = []
commenter1 = []
commenter2 = []
overlap = []
ratio1 = []
ratio2 = []

i = 1
k = 0
for channel in commenter_set_lst:
    j = i
    while j < len(commenter_set_lst):
        if j != i:
            channel1.append(channel_id[k])
            channel2.append(channel_id[j])
            commenter1.append(len(channel))
            commenter2.append(len(commenter_set_lst[j]))
            if commenter1 > commenter2:
                overlaps = len(channel) - len(channel - commenter_set_lst[j])
                overlap.append(overlaps)
            else:
                overlaps = len(commenter_set_lst[j]) -  len(commenter_set_lst[j] - channel)
                overlap.append(overlaps)
            if len(channel) == 0:
                ratio1.append(np.nan)
            else:
                ratio1.append(overlaps/len(channel))
            if len(commenter_set_lst[j]) == 0:
                ratio2.append(np.nan)
            else:
                ratio2.append(overlaps/len(commenter_set_lst[j]))
        j += 1
    i += 1
    k += 1
```

```python
channel1_df = pd.DataFrame(channel1)
channel2_df = pd.DataFrame(channel2)
commenter1_df = pd.DataFrame(commenter1)
commenter2_df = pd.DataFrame(commenter2)
overlap_df = pd.DataFrame(overlap)
ratio1_df = pd.DataFrame(ratio1)
ratio2_df = pd.DataFrame(ratio2)
```

```python
summary_df = pd.concat(
    [channel1_df, channel2_df, commenter1_df, commenter2_df, overlap_df, ratio1_df, ratio2_df], axis="columns", join="outer"
)
```

```python
summary_df
```

```python
diff_commenters = []
i = -1
for channelid in commenter_set_lst:
    i += 1
    single_lst = []
    for j in range(len(commenter_set_lst)):
        if j != i:
            diff = len(channelid) - len(channelid - commenter_set_lst[j])
            single_lst.append(diff)
        else:
            single_lst.append(np.nan)
    diff_commenters.append(single_lst)
```

```python
diff_commenters_df = pd.DataFrame(diff_commenters)
diff_commenters_df
```

```python
concat_df = pd.concat([channel_df, diff_commenters_df], axis ='columns', join='outer')
concat_df
```
